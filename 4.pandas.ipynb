# CSV

```
from io import StringIO, BytesIO
```

```
data=('c1,c2,c3\n'
'x,y,1\n'
'a,b,2\n'
'c,d,3')
```

```
type(data)
```

```
df1=pd.read_csv(StringIO(data))
```

```
df1
```

```
#specifying only certain columns 
df2=pd.read_csv(StringIO(data), usecols=['c1','c2'])
```

```
df2
```

```
#specifying different data types to columns
df=pd.read_csv(StringIO(data), dtype=object)
```

```
df['c1']
```

```
data=('index,c1,c2,c3\n'
'1,x,y,1\n'
'2,a,b,2\n'
'3,c,d,3')
```

```
#assigning index values
df=pd.read_csv(StringIO(data),index_col=0)
```

```
df
```

```
#Escape character
data= 'a,b\n"hello, \\"bob\\", nice to see you",5'
pd.read_csv(StringIO(data), escapechar='\\')
```
